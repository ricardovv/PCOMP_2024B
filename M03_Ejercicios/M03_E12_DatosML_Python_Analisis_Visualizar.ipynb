## E- VISUALIZAR SCATTERPLOT 4 CONJUNTOS EN UN PLOT (1 punto):

En este scaterpot graficaremos dos conjuntos de datos, I y II.
como ejercicio puede incorporar las  series III y IV y adaptar colores y titulos.

```python
# E1- Scatterplot oi grafico de puntos (pueden tener otras formas tb)

# Datos - - - - - - - - - - - -
# guardamos datos en variables
g1 = (df['I_x'], df['I_y'])
g2 = (df['II_x'], df['II_y'])
g3 = (df['III_x'], df['III_y'])
g4 = (df['IV_x'], df['IV_y'])

gdatos   = (g1, g2, g3, ? ) # guardamos esos grupos en la variable gdatos para acceder facilmente.
gnombres = ('El grupo 1', 'El grupo 2', 'El grupo 3', 'El grupo 4') # guardamos nombres de grupos
gcolores = ('red', 'blue', 'green', 'orange') # guardamos colores, puede cambiarlos: otors colores: https://matplotlib.org/stable/gallery/color/named_colors.html
gmarker  = ('o', 'v', 'd', 's'  ) # markers o formas de puntos, aca mas: https://matplotlib.org/stable/api/markers_api.html


### Crear gráficos - - - - - - - - - - - -
fig=plt.figure(figsize=(10, 6), facecolor='lightslategray') # tamaño del gráfico.
fig.patch.set_facecolor('green') # color del margen.
fig.patch.set_alpha(0.2) # transparencia del margen.

ax = fig.add_subplot(1, 1, 1) # agregar sub gráficos en el mismo espacio.
ax.patch.set_color('yellow') # color del gráfico.
ax.patch.set_alpha(0.9) # transparencia del gráfico.

plt.title("El Famoso Cuarteto de Anscombe") # titulo general.
plt.xlabel("Este es el Eje X") # texto eje X.
plt.ylabel("Este es el Eje Y") # texto eje Y.
plt.grid( color='grey', alpha=0.5) # grilla.

# se recorre la info de gdatos, gcolores, gnombres y se guardan en dat, col, gru.
for dat, col, gru, mar in zip(gdatos, gcolores, gnombres, gmarker):
    x, y = dat
    ax.scatter(x, y, alpha=0.6, c=col, edgecolors='none', s=80, label=gru, marker=mar) # puntos.
    plt.legend(gnombres) # leyendas.


# Mostrar grafico - - - - - - - - - - - -
# SOLO COMPU LOCAL guarda el grafico en la ruta seleccionada.
plt.savefig('a_cuarteto_scatter_series.jpg') # le piuede aparecer en la columna de al lado en los archivos

# DESCOMENTAR SOLO EN COLAB
#files.download("a_cuarteto_scatter_series.jpg")

# esta linea muestra el grafico.
plt.show()
```

```python
# F.1- Seleccionar articulo wikipedia

# Selecciona el idioma del artículo de WP.
wikipedia.set_lang("es")

# Ejemplo
# cual articulo a escoger: Visitar esta URL de ejemplo: https://es.wikipedia.org/wiki/La_Araucana
ejemplo_articulo_wp = 'La_Araucana'
articulo = wikipedia.page( ejemplo_articulo_wp )

## De las siguientes descomentar la que desee ver.

#print(articulo.original_title) # titulo del articulo
print( articulo.summary ) # sumario del articulo
print(articulo.content) # contenido textual del articulo
#print( articulo.categories ) # categorias del articulo
#print( articulo.images ) # lista url de imagenes del articulo
#print( articulo.url ) # url del articulo
#print(articulo.links[:]) # links que aparecen en el articulo
#print(articulo.html) # muestra html del articulo
```
