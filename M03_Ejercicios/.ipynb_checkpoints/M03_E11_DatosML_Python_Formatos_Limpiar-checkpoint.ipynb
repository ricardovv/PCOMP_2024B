```python
# 0.1- importar librerías.
import pandas as pd
import matplotlib.pyplot as plt
```

## 3- LIMPIEZA DE DATOS (2 puntos): 

Al inspeccionar, detectamos algunos problemas: muchos valores vacíos o NaN (detectar la fila con más NaN, o Not a Number), otros mal escritos (incosistencia de nombres y fechas), etc. Detectar los problemas permite definir las operaciones a realizar.   

En la etapa de limpieza por lo general se realizan una serie de operaciones:  
* formatear los datos unificando valores con puntos o comas
* se detecta informacion incompleta que en ciertos casos se completa o elimina. 
* se colocan textos en baja, 
* dependiendo de los datos, del problema o del escenario, pueden requerirse otras operaciones. 

Se recomienda que el resultado de la limpieza se guarde en un archivo para ser usado en etapas siguientes. Realizaremos algunas operaciones. 

**Si necesitan botar FILAS vacias o que no sirven:**    
– Indicar que filas: [desde:hasta]  
df = df.drop( df.index[11:955] )    
df # siempre mostrar los resultados  

**Si necesitan botar COLUMNAS vacias o que no sirven:**    
Indicar una lista de que columnas botar: ["esta", "y esta"]   
df = df.drop(['Unnamed: 6', 'Unnamed: 7'], axis=1)   
df # siempre mostrar los resultados    



```python
# 3.1- Con esta línea puedes botar una fila que no sirva o tenga muchos valores NAN
# en el ejemplo botaremos la fila 5 pues tiene muchos elementos nulos. 
# ojo: no correrla muchas veces, sino botar más filas. 
# De ser asi, repetir los pasos anteriores nuevamente, desde la carga de datos.

# OJO: asegurarse de botar la fila con mas NaN. En el ejempñlo la fila 5 
df = df.drop( df.index[ 5 ] )  
df
```

```python
# 3.2- Fijarse en el index
# Si al botar una o más filas, el index queda con números desordenados o falta de alguno 
# Se puede resetear el index con estas líneas

df.reset_index(inplace = True)
df
```

```python
# 3.3- Para continuar la limpieza, vamos a copiar df en otro dataframe llamado dflimpiando.
# Esto para ir separando los procesos.  
# nos crea otra columna de index que podemos borrar si lo estimamos. 

df = df.drop(['index'], axis=1)   
dflimpiando = df.copy()

# mostrar dataframe 
dflimpiando
```

```python
# 3.4- En la ultima columna del ejemplo hay una o más celdas que sin un valor el precio. 
# Podemos agregar un valor de la misma columna. agrgaremos 135
# asegurarse que sea la fila y columna correctas

dflimpiando.at[7, 'costounidad'] = 135

# mostrar dataframe 
dflimpiando
```

```python
# 3.5- Ahora veremos como unificar valores para que sean consistentes.  
# Se identifican los valores únicos en una columna: nombre e item
# para ver los valores mal escritos o inconsistentes y así poder unificar. 

dflimpiando['item'].unique()  
```

```python
# 3.6- Con esa información, vemos que en dicha columna hay inconsistencias: 
# 'LAPI', 'LAPICS', 'lapiz', deben ser 'lápiz'
# 'LIBRO', 'libro', deben ser 'libro'
# Por ahora reemplazaremos "a mano" valores inconsistentes por los correctos 
# Para esto usatmremos un diccionario con palabras 'incorrecta' : 'correcta'

dflimpiando = dflimpiando.replace( {'item':
                                    { 'LAPI':'lápiz',
                                      'LAPICS':'lápiz',
                                      'lapiz':'lápiz',
                                      'LIBRO':'libro'}
                                   } )
 
# mostrar dataframe 
dflimpiando
```

```python
# 3.7- Generalmente se unifican los nombres, y se ponen en bajas
# Seleccionar la columna a unificar y poner valores en bajas con las funciones .str.lower() 

dflimpiando['nombre'] = dflimpiando['nombre'].str.lower() 

# mostrar dataframe 
dflimpiando
```

```python
# 3.8- Las fechas están en diferentes formatos, se debe unificar. 

dflimpiando['fecha'] = pd.to_datetime(dflimpiando.fecha)

# mostrar dataframe 
dflimpiando
```

# 4- DATOS DERIVADOS (1 punto): 

Los datos ahora estan en mejor forma, pero podemos crear datos nuevos, derivándolos de datos existentes. En el ejemñplo, podemos hacer una columna derivada con el total de los valores. 


```python
# 4.1- Crear una nueva columna en base a otras (de la cantidad y el precio unitario) 
# total = unitario x cantidad 
# dflimpiando['total'] crea la columna nueva con el resultado de la operación entre otras dos columnas:  

dflimpiando['total'] = dflimpiando['cantidad'] * dflimpiando['costounidad']

# mostrar dataframe 
dflimpiando
```

```python
# 4.2- Ahaora tenemos los datos algo más ordenado y prácticamente limpios. 
# Mostramos el dflimpiando para ver como va quedando.

dflimpiando 
```

# 5- PERSISTENCIA DE DATOS: GUARDAR ARCHIVOS (1 punto): 

Si estamos ok con el archivo dflimpiando, lo guardamos en el computador para trabajar en etapas posteriores como el análisis usando este nuevo archivo limpio (que después debemos cargar para trabajar con este).

De esta manera vamos separando los procesos. Limpieza es un proceso que termina con un archivo nuevo, y con este archivo nuevo se inicia la etapa de análisis, que será breve en este ejercicio. 


```python
# 5.1- Guardar datos limpios para usarlos en los procesos siguientes. 
# Puede guardar alguno o todos los formatos para ver su estructura. 
# Los mas complejos con XML y JSON pues debe especificarse su estructura. 
# Recordar que si los uarda desde colab debe cambiar la ruta y nombre del archovo segun correpwonda (libros, peliculas, etc)
# Averiguar como podria guardar en local desde colab.  

ruta_save = 'e11_datos/'

# Para guardar como CSV
dflimpiando.to_csv( ruta_save + 'ciej_utiles_limpio.csv' )

# Para guardar como TSV, igual que csv solo cambia la extensión y se agrega separador
dflimpiando.to_csv( ruta_save + 'ciej_utiles_limpio.tsv', sep="\t" )

# Para guardar como EXCEL
dflimpiando.to_excel( ruta_save + 'ciej_utiles_limpio.xlsx' )

# Para guardar como JSON
dflimpiando.to_json( ruta_save + 'ciej_utiles_limpio.json' )

# Para guardar como XML
dflimpiando.to_xml( ruta_save + 'ciej_utiles_limpio.xml' )

```

# 6- VISUALIZACIÓN: BARAS Y SCATTERPLOT (1 punto):  

Una parte importante del análisis es la visualización de datos. No sirve solo para mostrar los datos, sino que es usada para explorar desde una perspectiva visual como se comportan, relacionan o distribuyen los datos.  


```python
# 6.1- Traer datos limpios que guardamos anteriormente.
# Debe entonces importar desde su compu (o la nube si esta en colab). 

dflimpio = pd.read_csv('e11_datos/ciej_utiles_limpio.csv') 

# mostrar dflimpio
dflimpio
```

```python
# 6.2- Eliminar columna 'Unnamed: 0' para dejar solo lo que nos sirve.

dflimpio.drop(['Unnamed: 0'], axis='columns', inplace=True) # inplace para guardar en el mismo dataframe. 

# mostraremos sin columna borrada. 
dflimpio 
```

```python
# 6.3- Con los datos del ejemplo, Haremos un gráfico de barras simple para mostrar algunos datos.

# Guardamos las columnas en unas variables para que sea más fácil. 
valor1 = dflimpio['fecha']
valor2 = dflimpio['cantidad']

# Hacer el gráfico
plt.figure( figsize=(14,4) ) # ancho x alto 
plt.bar(x=valor1, height=valor2)
plt.show()  

# Guarda el gráfico en la ruta seleccionada.
#plt.savefig('e11_datos/ciej_utiles_limpio_barra1.jpg') #descomentar para guardar local

# Si esta en colab puede usar algo como esto. Descomentar para usar
#from google.colab import files
#plt.savefig("migrafico1.png")
#files.download("migrafico1.png") 
```

```python
# 6.4- Ahora probaremos con gráfico de puntos.

# Guardamos las columnas en unas variables.
v1 = dflimpio['item']
v2 = dflimpio['cantidad']

fig=plt.figure(figsize=(8, 4)) # tamaño del gráfico. 

plt.title("COMPRAS UTILES; dos variables") # título. 
plt.xlabel("Este es el Eje X") # texto eje X. 
plt.ylabel("Este es el Eje Y") # texto eje Y. 
plt.scatter(x=v1, y=v2, color='red') # gráfico de puntos.

# esta línea muestra el gráfico.
plt.show() 
```
