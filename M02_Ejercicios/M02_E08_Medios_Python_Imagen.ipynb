## A- Imagen: Imagen es una matriz. 

Para empezar considerar que una imagen es una matriz de valores. Veamos la diferencia entre vactor y matriz: 

**Vector:** Es una lista de elementos, como las liastas que hemos visto en python, se extiende en una dimensión, por ejemplo: 
```  
mivector = [ 56, 7, 34, 9, 23 ] # lista simple de 6 elementos.  
```  
**Matríz:**  Es una lista de elementos pero que se extiende en dos dimensiones, como una cuadrícula, por ejemplo: 

```  
mimatriz = [ [9, 24, 7], [13, -5, 42], [81, 43, 2] ]  # matriz de 3x3 elementos.  
```  
Esa matriz se vería algo así: 
```  
[
[ 9, 24, 7 ], 
[ 13, -5, 42],
[ 81, 43, 2]  
]
```  
Entonces las imágenes son lo mismo, una matriz de valores (de grises, rgb, Hexadecimal o el espacio que sea) 
Al ser una matriz, se pueden realizar muchas operaciones de transformación (escalar, trotar, etc). Se pueden modificar pixeles. Todo eso es lo que en el fondo hacen programas como photoshop, modificar pixeles aplicando transformaciones a los valores y a la matriz.   


````python
# A- Imagen: Imagen es una matriz. Ejemplo matríz de valores como imagen.  

# Con esto hacemos una matriz de 5x5, la llenamos con números aleatorios. 
# valores entre 50 y 255, 10 filas x 10 columnas. 
img_random = np.random.randint(10, 255,size = (5,5))

# mostramos valores de esta matriz. 
img_random
````

````python
# A- Imagen: Imagen es una matriz. Ejemplo: 
# Ahora mostramos esta matriz de números, como colores
# Los valores numéricos de la matriz no tienen color en si mismos, deben "tomar" un tipo de color. 
# Esta es una lista de posibles colores en MatPLotLib. 
# Ojo, esta es la matriz de UN solo canal, y por lo general trabajaremos con imagenes RGB, de 3 canales. 

# Estos son "Colores prefabricados", podríamos decir, que usaremos para dar color a los valores de la matriz.   
colores_mapeo = ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                 'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                 'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']


# Mostrar la matriz con Matplotlib
fig, axrand = plt.subplots(figsize=(3, 3)) # tamaño
axrand.set_title('Imagen matriz c valores random') # titulo
axrand.imshow(img_random, cmap=colores_mapeo[8] )  # mostrar imagen, cambiar valor en colores_mapeo[n]
#axrand.axis('off') # descomentar para sacar los ejes 
````

````python
# A- Imagen: Imagen es una matriz. Ejemplo Pintar solo un pixel 

# Crearemos otra imagen desde cero, tamaño 50px x 50px, con todos los pixeles negros. 
img_fome = np.zeros((50, 50, 3), np.uint8)

# Colores para pixeles
fred = [ 255,0,0 ]
fgrn = [ 0,255,0 ]
fblu = [ 0,0,255 ]

# Damos los valores a los pixeles en sus posiciones x e y 
img_fome[3,5] = fred # el pixel de la posicion 4,5
img_fome[16,27] = fgrn # el pixel de la posicion 6,7
img_fome[38,41] = fblu # el pixel de la posicion 8,9


# Con un ciclo tb se peden asignar valores.
for i in range( 20 ):
    img_fome[20+i,10+i] = fgrn 
    
    
# Moatrar la imagen con los pixeles coloreados-. 
plt.imshow(img_fome)
plt.show()
 
````

## C- Imagen: Operaciones. 

Al ser números, se pueden realizar operaciones de transformación a la imagen, es decir, a la matríz. Estas transformaciones son muy conocidas en diseño: desplazar, rotar, escalar (ampliar-reducir), etc. Todas operan en la matriz mediante operaciones matemáticas, que afortunadamente ya vienen definidas en algunas librerías de python. 


````python
# C- Imagen: Operaciones. Ejemplo Rotación. 

# Rotar matriz. 
fig, axflor = plt.subplots(figsize=(5, 5)) # tamaño
tr = transforms.Affine2D().rotate_deg( 19.9 )
axflor.imshow(imgflor, transform=tr + axflor.transData)
#axflor.axis('off')
plt.show()
````

````python
# Recortar 
# para recortar tienen que ver cual es la forma de la imagen, para estar dentro de los parametros de ancho y alto
# sabemos que imgflor tiene 250x250 pixeles. 

imgflor_recortada = imgflor[ 0:150, 0:100 ] # alto-ancho
plt.imshow(imgflor_recortada)
plt.axis("off")
plt.show()
````

## D- Pixeles: Valores de Pixeles.

Se pueden cambiar los valores globales de como se "lee" el valor de pixel, lo que afecta como se mostrará una imagen. Para esto se puede usar la librería MatPlotLib (que debe cargarse al inicio) y seleccionar los espacios de color que interesan. 

También se puede seleccionar mostrar cada uno de los canales de la imagen. Por ejemplo si la imagen es RGB, se puede seleccionar mostrar el canal r, g, o el b.

## F- Pixeles: Visualizar distribución con Histograma. 

Las imágenes tienen ancho y alto, lo que da una cantidad de pixeles de la imagen (alto x ancho) y cada uno de estos pixeles puede tener un valor en RGB. A veces es pertinente mostrar en una visualización para ver como se distribuyen los valores de pixeles, por ejemplo para ver la predominancia de ciertos colores en una imagen. Para esto se usa un gráfico llamado histograma, que cuenta cuantas veces se repite un valor.  


````python
# Visualizar distribución con Histograma.
# Por ejemplo el 150 aparece 10 veces en capa R, el 25 aparece 90 veces en capa G, etc...

# Se selecciona cada uno de los 3 canales. 
imgflor_cR = imgflor[:, :, 0] # [ todos los pixeles del r, todos los pixeles del g, cual canal ]
imgflor_cG = imgflor[:, :, 1]
imgflor_cB = imgflor[:, :, 2]

# Cada acanal (RGB) se aplana de matriz 2d (anchoxalto), a un vector 1D. 
red_pxls = imgflor_cR.flatten()
gre_pxls = imgflor_cG.flatten()  
blu_pxls = imgflor_cB.flatten()  


# Para mostrar histograma de la imagen. 
# proporción del gráfico 
plt.figure(figsize=( 16, 6 ) )

# que es cada parametro: plt.hist(cual canal, bins=cuantas barritas, density=False, color=color de las barritas, alpha=transparencia de las barras)
plt.hist(red_pxls, bins=256, color='red',   alpha=0.4)
plt.hist(gre_pxls, bins=256, color='green', alpha=0.4)
plt.hist(blu_pxls, bins=256, color='blue',  alpha=0.3)

# Titulos
plt.title('Histograma Colores')
plt.ylabel('Cuantos por cada valor')
plt.xlabel('Intensidad de los valores de 0 a 255')

# Mostrar
plt.show()
````
